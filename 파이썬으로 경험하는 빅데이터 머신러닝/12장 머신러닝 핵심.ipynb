# 머신러닝

## 프로그래밍과 머신러닝
- 머신 러닝은 데이터로부터 결과를 찾는 것이 아니라, 주어진 데이터로부터 규칙성을 찾는 것이 묙포입니다.
<br><br>
- 일단 규칙성을 발견(이 규칙성을 머신러닝 모델이라고 함)하면 이 후에 들어오는 새로운 데이터에 대해서 발견한 규칙성을 기준으로 정답을 찾아내는데, 이는 기존의 프로그래밍 방식으로 접근하기 어려웠던 문제의 해결책입니다.
<br><br>
- 머신러닝이 기존 프로그래밍의 한계를 해결할 수 있는 문제는 문제 해결 접근 방법이 기존 프로그래밍 방식과는 다르기 때문입니다.
<br><br>
|문제|사람|기존프로그램|머신러닝|
|:------:|:---:|:---:|:--:|
|수천 개의 숫자를 빠르게 곱하기|어려움|쉬움|쉬움|
|사진에서 사람들의 얼굴 찾기|쉬움|어려움|쉬움|

## 머신러닝의 발전
- 일반적인 주장은 1949년 캐나다 심리학자 햅(Hebb)이 햅학습 이론(Hebbian Learning Theory)를 발표하는 것으로 시작되었습니다.
<br><br>
- 이후 1952년 IBM서 근무하던 아사새뮤얼(Arthur Samuel)은 최초의 머신러닝 프로그램이라 할 수 있는 체커 프로그램을 개발하였습니다.
<br><br>
- 이때 경험으로부터 배우는 방법을 이론을 주장하였고 여기서 머신러닝이란 '명확히 프로그램하지 않고도 컴퓨터에 사고하는 기능'이라고 정의하였습니다.

# 머신러닝 구성

## 머신러닝 데이터
- 많은 머신 러닝 문제가 1개 이상의 독립 변수 X를 가지고 종속 변수 Y를 예측하는 문제입니다.
<br><br>
- 특히 딥러닝과 같은 인공 신경망 모델은 독립 변수, 종속 변수, 가중치, 편향 등을 행렬을 통해 연산하는 경우가 많습니다.
<br><br>
- 딥러닝을 배우게 되면 훈련 데이터를 행렬로 표현하는 경우를 많이 보게 되는데 판다스에서 배우는 데이터프레임 구조를 행렬로 이해하면 됩니다.
<br><br>
- 머신러닝에서는 하나의 데이터, 하나의 행을 샘플(Sample)이라 부르고 종속변수를 예측하기 위한 각각의 독립변수를 특징(Feature)이라고 합니다.

## 머신러닝 문제
- 머신러닝 문제들은 대부분 분류와 회귀에 속합니다.
<br><br>
- 예(Y)/아니오(N) 등과 같은 형태입니다.
<br><br>
<br><br>
- 회귀는 연속된 값을 예측하는 것입니다.
<br><br>
- 정리하면 분류는 범주형을 예측하는 것이고 회귀는 연속형을 예측한다고 생각하면 됩니다.

### 분류
- 분류(Classification)는 단어 그대로 클래스(라벨, 종속변수)을 구분하여 예측하는 것입니다.
<br><br>
- 예측 해야할 대상 즉, 클래스가 정해져 있습니다.
<br><br>
- 분류는 또한 이진 분류(Binary Classifcation)과 다중 클래스 분류(Multi-Class Classfication)로 구분 됩니다.

### 이진 분류
- 이진 분류(Binary Classification)는 주어진 독립 변수에 대해서 종속변수가 둘 중 하나의 답을 구분하는 문제입니다. 앞서 로지스틱회귀에서 점수에 따른 시험 성적에 대해서 합격, 불합격 여부를 판단하는 문제가 여기에 속하며 받은 메일이 정상 메일, 스팸 메일인지를 판단하는 문제등도 이진 분류에 속합니다.

### 다중 클래스 분류(Multi-class Classification)
- 다중 클래스 분류는 입력 변수 대한 두 개 이상의 정해진 선택지 중에서 답을 분류하는 ㅁ누제입니다. 머신러닝 사례로 많이 소개한 붓꽃의 품종을 예측하는 것은 3개의 클래스를 가지고 잇기 때문에 다중 분류에 속합니다. 3개의 선택지를 주로 카테고리 또는 범주 또는 클래스라고 하며, 주어진 입력으로부터 정해진 클래스 중 하나로 판단하는 것을 다중 클래스 분류 문제라고 합니다.

## 회귀
- 회귀(Regression)은 분류 문제처럼 0 또는 1이나 붓꽃의 클래스와 같이 분리된(비연속적인) 답을 예측하는 것이 아니라 연속된 값을 예측합니다.
<br><br>
- 회귀는 연속적인 숫자(실수)를 예측합니다.

# 머신러닝 학습구분

## 머신러닝 학습
- 머신러닝 과정은 데이터를 사용해 모델을 생성하고 학습을 반복하는 과정입니다.
<br><br>
- 머신러닝에서 학습은 지도 학습, 비지도 학습, 강화 학습으로 구분합니다.

### 지도학습
- 지도 또는 교사학습(Supervised learning)은 어린 아이의 학습과정을 살펴보면 알 수 있습니다. 부모가 선생님(Supervisor) 역할을 하며 시행 착오를 통하여 오차를 최소화하는 방식입니다. 지도학습은 대량의 훈련 데이터를 선생님과 같이 기계에서 학습 시키고, 결과를 조사하고, 기대한 결과를 얻을 때까지 매개변수를 조정합니다. 그런 다음 이전에 본 적이 없는 일련의 검증 데이터에 대해 결과를 예측하도록 함으로써 무엇을 배웠는지 알 수 있게 되는 방법입니다. 여기서 선생님 역할을 하는 변수가 레이블(클래스, 종속변수)입니다.

### 비지도 학습
- 비지도 학습은 레이블 없이 학습하는 것을 의미합니다. 여기서 레이블이 없다는 의미는 정답을 따로 알려주지 않는다는 의미로 비슷한 데이터들을 군집화하는 방법입니다. 예를 들어 호랑이, 코끼리, 사자, 얼룩말 사진을 비지도학습으로 머신러닝 모델을 만든다면 각 사진이 무슨 동물인지 정답(레이블)을 알려주지 않았기 때문에 이 동물이 '무엇'이라고 기계가 정의할 수 없기 때문에 비슷한 무리로 군집화하는 방법으로 기계가 학습합니다.

|구분|지도 학습|비지도 학습|
|:---:|:---:|:---:|
|목적|예측 모델 생성|고차원 데이터 분류|
|장점|인간이 결과값에 개입하여 정확도가 높음|결과 값을 정해주지 않아도 되므로 속도 빠름|
|단점|시간이 오래 걸리고 학습 데이터 양이 많음|학습 결과로 분류 기준과 군집 예측 불가하며 분석가의 판단에 의지|
|검증방법|교차 검증 수행|별도 검증 없음|
|데이터|라벨링 데이터 / 클래스 데이터|원 데이터|
|유형|회귀: Y = X / 분류: 그룹별 특징 파악|군집: 데이터 클러스터링/ 패턴인식: 여러그룹인식|
|주요알고리즘|CNN, RNN, SVM. DT|K-Means, DBSCAN, Clusturing|
|사례|패턴인식, 질병진단|스팸필터, 차원 축소|

### 강화학습
- 강화학습(Reinforcement Learing)은 상과 벌이라는 과정에서 보상(reward)을 통해 상을 최대화하고 벌을 최소화하면서 학습하는 방식
<br><br>
- 유명한 알파고가 이 방법을 사요
<br><br>
- 정확한 정담은 없지만, 반복적인 행동들을 통해 칭찬과 벌을 받음으로써 보상의 가중치를 최대화하는 것이 강화학습 방법

## 머신러닝 평가

### 데이터분할
- 머신러닝 모델이 완성되었다면 전체 데이터를 학습을 위한 훈련용, 평가를 위한 테스트용 두 가지로 분할(Split)합니다.
<br><br>
- 특별한 경우 세 가지 훈련용, 검증용, 테스트용으로 분리해서 하기도 합니다.
<br><br>
>- 훈련데이터(Training data): 머신러닝 모델을 학습하는데 사용하는 데이터이며 모델이 열심히 공부하는데 과정에 사용하는 데이터
<br><br>
>- 검증데이터(Validation data): 모델 훈련에 적절한 지점을 찾기 위해 구성하는 데이터, 필수는 아니며 최종 시험을 보기 전에 모의고사 문제를 보는 데이터
<br><br>
>- 테스트 데이터(Test data): 모델의 성능을 평가하기 위해 사용하는 데이터, 이 데이터를 통해 머신러닝 모델이 얼마나 잘 학습되었는지 평가하는데 사용

### 모델 평가
- 모델 평가는 우수한 머신러닝 모델을 결정하는 단계로 얼마나 정확하게 예측한 결과와 실제 정답이 일치하는 지를 검증하는 과정
<br><br>
- 많이 사용하는 방법으로 혼동행렬(Confusion matrix)이 있음.
<br><br>
- 홍동행렬은 정답이 True와 False로 나누어져 있고, 아래와 같이 2x2 매트릭스로 구분
|예측/정답|참(True)|거짓(False)|
|------|---|---|
|참(True)|TP|FP|
|거짓(False)|FN|TN|
<br><br>
1. 정확도
>- 정확도(Accuracy)는 머신러닝에서 가장 대표적으로 사용되는 지표로 전체 데이터중에서 정확히 분류된 데이터의 비율이 얼마나 되며 모델을 정확하게 분류 하는지를 나타내는 지표입니다. (수식: TP+TN / TP+FN+FP+TN)
<br><br>
1. 정밀도
>- 정밀도(Precision)는 실제 해당 데이터 중 모델이 올바르게 검출한 데이터의 비율입니다. (수식: TP / TP+FP)
<br><br>
1. 재현률
>- 재현률(Recall)은 실제값이 정답인 데이터 전체 개수에 대해 TP의 비율 즉, 정답인 데이터 중에서 얼마나 정답인지를 예측(재현)했는지를 나타내는 지표입니다.(수식: TP / TP+FN)
<br><br>
1. ROC
>- ROC(Receiver Operating Characteristic) 학습 곡선은 y축에 TP, x축에 FP 수치를 배치해 두 수치의 균형을 살펴보는 머신러닝 평가를 위한 시각화 모델입니다.
<br><br>
>- ROC 커프 아래 면적인 AUC(Area Under Curve)가 1에 가까어질 수록, 모델이 Y를 예측하는 정확도가 높은 모델입니다.

### 케라스
- 케라스(Keras)는 대부분 종류의 딥러닝 모델을 간편하게 만들고 훈련시킬수 있는 파이썬 딥러닝 프레임워크입니다.
<br><br>
- 케라스는 백엔드 엔진으로 텐서라이브러리를 사용하며 딥러닝 모델을 위한 고수준의 구성 요소를 제공하는데, 텐서플로와 같은 저수준의 연산은 다루지 않습니다.
<br><br>
- 케라스는 모듈 구조로 구성되어 있어 하나의 텐서라이브러리에 국한하여 구현되어 있지 않고, 여러가지 백엔드 엔진과 자연스럽게 연동됩니다.
<br><br>
- 현재는 텐서플로우, 티아노, CNTK 3개를 백엔드 엔진으로 사용하고 있습니다.
<br><br>
- 케라스 특징

<br><br>
>- 동일한 코드로 CPU와 GPU에서 실행할 수 있습니다.
<br><br>
>- 사용하기 쉬운 API로 딥러닝 모델 프로토타입을 빠르게 만들 수 있습니다.
<br><br>
>- 컴퓨터 비전을 위한 CNN, RNN을 지원하며 자유롭게 조합하여 사용할 수 있습니다.

## 로지스틱 모델 예측
- 지금부터는 머신러닝 과정을 통해 타이타닉 생존자를 예측 해보겠습니다.

```python
import pandas as pd
from sklearn import model_selection
from sklearn import metrics
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# 타이타닉 데이터 준비
file_path = 'C:/Users\BurgerC/파이썬으로 경험하는 빅데이터 머신러닝/data/titanic.csv'

#read_csv() 함수로 데이터프레임 변환
df = pd.read_csv(file_path)
df

## 데이터 전처리
# 불필요 컬럼 삭제 (x 변수)
del df['PassengerId']
del df['Name']
del df['Ticket']
del df['Cabin']

#결측치 처리 (x)
df.dropna(thresh = int(len(df) * 0.5), axis = 1)  # 결측치 제거
df['Embarked'] = df['Embarked'].fillna('S')  # 최다빈도 'S'로 대체
df['Age']=df['Age'].fillna(df['Age'].mean())  # 나이 평균값으로 대체

# 범주형 컬럼 처리 (x)
df['Sex'] = df['Sex'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df = pd.get_dummies(df)

# 독립변수와 종속변수 구분 (x, y)
x_data = df.iloc[:, 1:]
y_data = df.iloc[:,0]
x_data = x_data.values
y_data = y_data.values
print(x_data)
print(y_data)

## 머신러닝 
#데이터분할(7:3)
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size=0.3)

#로지스틱 모델링 
estimator = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)
estimator.fit(x_train, y_train)

# train 평가 
y_predict = estimator.predict(x_train)
score = metrics.accuracy_score(y_train, y_predict) #classification
print('train score: ', score)

# test 평가 
y_predict = estimator.predict(x_test)
score = metrics.accuracy_score(y_test, y_predict)
print('test score: ', score)


# 모델 적용
print(x_test[:2])
y_predict = estimator.predict(x_test[:2])
print(y_predict)
for y1, y2 in zip(y_test, y_predict):
    print(y1, y2, y1==y2)
```

- 모델 평가를 보면 학습단계 훈련데이터의 정확도는 81.8%인데 이를 테스트데이터에 적용할 경우 다시 떨어진 76.8%의 정확도를 확인할 수 있습니다.

## 의사결정나무 예측

- 의사결정나무(Decision Tree)는 여러 가지 규칙을 순차적으로 적용하면서 독립 변수 공간을 분할하는 분류 모형입니다.
<br><br>
- 분류와 회귀 분석에 모두 사용될 수 있기 때문에 CART(Classification And Regression Tree)라고도 합니다.
<br><br>
- 의사결정나무는 예/아니오 질문을 이어가며 학습을 수행합니다.
<br><br>
- 한 번의 분기 때마다 변수 영역을 두 개로 구분하며 결정 트리에서 질문이나 정답을 담은 네모 상자를 노드(Node)라고 합니다.
<br><br>
- 맨 처음 분류 기준(즉, 첫 질문)을 뿌리노드(Root Node)라고 하고, 맨 마지막 노드를 종료노드(Leaf Node)라고 합니다.

```python
import pandas as pd
from sklearn import model_selection
from sklearn import metrics
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# 타이타닉 데이터 준비
file_path = 'C:/Users\BurgerC/파이썬으로 경험하는 빅데이터 머신러닝/data/titanic.csv'

#read_csv() 함수로 데이터프레임 변환
df = pd.read_csv(file_path)
df

## 데이터 전처리
# 불필요 컬럼 삭제 (x 변수)
del df['PassengerId']
del df['Name']
del df['Ticket']
del df['Cabin']

#결측치 처리 (x)
df.dropna(thresh = int(len(df) * 0.5), axis = 1)  # 결측치 제거
df['Embarked'] = df['Embarked'].fillna('S')  # 최다빈도 'S'로 대체
df['Age']=df['Age'].fillna(df['Age'].mean())  # 나이 평균값으로 대체

# 범주형 컬럼 처리 (x)
df['Sex'] = df['Sex'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df = pd.get_dummies(df)

# 독립변수와 종속변수 구분 (x, y)
x_data = df.iloc[:, 1:]
y_data = df.iloc[:,0]
x_data = x_data.values
y_data = y_data.values
print(x_data)
print(y_data)

## 머신러닝 
#데이터분할(7:3)
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size=0.3)

# 의사결정나무 모델링
estimator = DecisionTreeClassifier(criterion='gini', max_depth=None, max_leaf_nodes=None,
                                  min_samples_split=2, min_samples_leaf=1,
                                  max_features=None)
estimator.fit(x_train, y_train)

y_predict = estimator.predict(x_train)
score = metrics.accuracy_score(y_train, y_predict) #classification
print('train score: ', score)

# test 평가 
y_predict = estimator.predict(x_test)
score = metrics.accuracy_score(y_test, y_predict)
print('test score: ', score)

# 모델 검증
print(confusion_matrix(y_test, y_predict))
print(classification_report(y_test, y_predict))

# 중요도
estimator.fit(x_train, y_train)
print(estimator.feature_importances_)

d = {'attribute': df.iloc[:, 1:].columns, 'importance': estimator.feature_importances_}
df_importance = pd.DataFrame(d)
df_importance.sort_values('importance', ascending=False)
print(df_importance)

# 의사결정나무 그리기
# http://www.graphviz.org/download/
#pip install graphviz
```

- 모델 평가를 보면 훈련데이터의 정확도는 98%이고 이를 테스트데이터에 적용할 경우 76%의 정확도를 확인할 수 있습니다.
