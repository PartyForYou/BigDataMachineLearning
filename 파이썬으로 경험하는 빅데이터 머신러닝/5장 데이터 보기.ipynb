## 범주형
- 범주형 변수의 분석의 첫 단계는 해당 변수의 빈도(Frequency)와 전체 데이터 빈도를 계산하는 것부터 시작합니다.

### 빈도분석
- 단일 범주형 변수 기술 분석의 첫 번째 과정은 범주형 변수 값에 빈도를 파악하는 것입니다.
- 빈도는 대상 자료에서 반복하는 횟수를 기록하는 것으로 빈도분석을 통해 전반적 분포를 살펴보면서 전체 데이터의 구성을 파악하는 과정입니다.
- 가장 빈도가 높은 데이터와 가장 빈도가 낮은 데이터를 기록하고 데이터 분석을 시작해야 합니다.

### 교차분석
- 교차분석은 범주형 변수 간 빈도분석을 확장하여 빈도를 교차시킨 분할표를 만들어 분석하는 방법입니다.
- 분할표는 교차분석표(Cross table)라고도 합니다.

```python
# 관련 라이브러리 불러오기
import pandas as pd  # 판다스
import matplotlib as mpl  # 맷플롭립
import matplotlib.pylab as plt  # 맷플롭립
import seaborn as sns  # 시본
import numpy as np  # 넘파이

# 데이터 준비
tips = sns.load_dataset("tips")  # 팁 데이터

# 데이터 보기
tips
```

```python
# tips eㅔ이터 유형 확인
# tips 데이터의 범주형(category) 변수는 sex, smoker, day, time 4개

tips.dtypes
```

```python
# 성별 범주형 빈도분석
tips['sex'].value_counts()
```

```python
# 요일별 범주형 빈도분석
tips['day'].value_counts()
```

```python
# 성별, 요일별 교차분석
pd.crosstab(tips['sex'],tips['day'])
```

```python
# 여백 또는 누적값cumulatives
pd.crosstab(tips['sex'], tips['day'], margins=True)
```

```python
# 전체 빈도 비율 확인
pd.crosstab(tips['sex'], tips['day']).apply(lambda r: r/len(tips), axis=1)
```

## 연속형
- 연속형 분석은 수직된 수치 데이터의 정리, 표현, 요약 등을 통해 데이터의 전반적인 특성을 이해하는 분석 진입 단계입니다.
- '기술통계'라고도 합니다.

### 데이터의 중심 이해: 평균, 중위수
- 다양한 평균의 세계
> - 산술평균 : 모든 값을 더한 후 값의 개수만큼 나눈 후 나오는 값을 의미<br>
> - 기하평균 : %로 평균 비율을 구할 때 방법<br>
> - 제곱평균 : 각 값의 제곱의 평균을 구한 후 루트를 적용해서 구하는 평균<br>
> - 조화평균 : 평균 속도를 구할 때 주로 사용하는 방법

### 데이터 퍼짐 정도 이해 : 분산, 표준편차, 범위, 사분위수
  - 분산: 데이터 퍼짐 정도
  - 표준편차: 데이터의 퍼짐 정도를 나타내는 또 하나의 지표이고 동일한 기분을 적용하기 위한 편차 제곱합의 평균으로 나타냄
  - 범위: 데이터의 흩어진 범위
  - 사분위: 데이터의 분포가 좌우대칭이 아니거나 이상치가 있는 경유 평균은 극단적으로 치우진 대표성이 없는 값에 의해 영향을 받는다. 이 경우 자료를 나열하여 전체 데이터를 파악할 수 있습니다. 자료를 순서대로 나열했을 때 즉, 50%에 위치하는 수가 중위수(Q2)이고 25%에 위치하는 수가 Q1입니다.

## 데이터 병합
- 병합 명령은 두 데이터 프레임의 공통 열 혹은 인덱스를 기준으로 두 개의 테이블을 병합하는 과정입니다.
- 이 때 기준이 되는 열, 행의 데이터를 키(key)라고 합니다.
> df.merge()

```python
import pandas as pd

# 병합용 데이터프레임 생성
df1 = pd.DataFrame({'고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007], '이름':['강감찬', '홍길동' , '이순신', '장보고', '유관순', '신사임당', '세종대왕']}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({'고객번호': [1001, 1001, 1005, 1006, 1008, 1001], '금액': [10000, 20000, 15000, 5000, 100000, 3000]}, columns=['고객번호', '금액'])
df2
```

```python
# inner join
pd.merge(df1, df2)
```

```python
# Full join 방식은 키 값이 한쪽에만 있어도 데이터를 출력
pd.merge(df1, df2, how='outer')
```

```python
# left, right 방식은 각각 첫 번째, 혹은 두 번째 데이터프레임의 키 값을 모두 보여준다.
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how = 'right')
```
