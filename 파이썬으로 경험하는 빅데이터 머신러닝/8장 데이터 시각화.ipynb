# 시각화의 시대

- 데이터 시각화(Data visualization)는 매우 효과적으로 정보를 전달하는 수단이 되고 있다.

## 눈으로 보는 데이터

- 존스노우 콜레라 맵은 본질적으로 점을 활용한 맵 시각화입니다.

## 시각화의 중요성

- 단순 수치를 시각화된 그림이나 도표로 보면 한눈에 어떠한 요소들이 서로 더 관계가 있는지 쉽게 알 수 있습니다.
<br><br>
- 데이터 분석 시 수치에 의존한 기술 통계 의사결정뿐만 아니라 시각화를 통해 검증해야 한다는 데이터 시각화의 중요성을 다시 한 번 강조하고 있습니다.

## 시각화 라이브러리

>- 내장라이브러리 : 판다스에 내장된 기본 그래프 라이브러리를 별도의 라이브러리 임포트 없이 사용 가능합니다.<br><br>
>- 맷플롯립(Matplotlib) : 판다스에서 가장 많이 쓰는 라이브러리로 데이터프레임을 시각화 할 때도 내부적으로 맷플롯립을 사용합니다.<br><br>
>- 시본(Seaborn) : 시본은 맷플롯립을 기반으로 다양한 색 테마, 차트 기능을 추가한 라이브러리로 맷플롯립에 의존성을 가지고 있습니다. 맷플롯립에 없는 그래프(히트맵, 카운트플랏 등)을 가지고 있습니다.<br><br>
>- 플로트나인(Plotnine) : Plotninne은 R의 ggplot2에 기반해 그래프를 작성하는 라이브러리로 R로 시각화하는 것이 익숙하면 편리하게 사용할 수 있습니다.<br><br>
>- 폴리옴(Folium) : Folium은 지도 데이터(Open Street Map)에 leafletjs를 이용해 위치정보를 시각화하는 라이브러리로 자바스크립트 기반으로 인터랙티브하게 그래프를 그릴 수 있습니다. 한국 GeoJSON 데이터는 깃허브 southkorea-maps에서 확인할 수 있습니다.<br><br>
>- 파이차트(Pyecharts) : 중국 바이두에서 데이터 시각화를 위해 만든 Echarts.js의 파이썬 버전으로 다양한 그래프들이 내장되어 있어 레포트를 작성할 때 효과적입니다.<br><br>
>- 플로틀리(Plot.ly) : plotly는 인터랙티브 그래프를 그려주는 라이브러리로 R, 스칼라, 파이썬, 자바스크립트, 매트랩 드엥서 사용할 수 있습니다. 시각화를 위해 D3.js를 사용하고 있으며 사용해보면 사용이 쉽고, 세련된 도구이다 Online과 Offline이 따로 존재하며 온라인 이유 시 별도 APO key가 필요합니다. Plotly cloud 유로 모델도 있습니다.

## 시각화 준비

### 일반적인 시각화 단계
1. 1단계

## 시각화 옵션 지정

### 제목
- 제목은 title() 함수를 추가합니다.

```python
# 제목추가
import matplotlib.pyplot as plt
plt.title('Line')
plt.plot([1, 2, 3], [1, 2, 3], marker='o')
plt.show()
```

### 범례
- 범례(legend)는 두 개 이상의 데이터를 표시할 때 사용합니다.
<br>
<br>
- 범례를 추가하기 위해 plot() 함수를 응용합니다.
<br>
<br>
- Plot()함수에 label이라는 속성 레이블값으로 원하는 문자열을 추가하고 그래프를 그리기 전에 legend() 함수를 실행합니다.

```python
# 범례추가
plt.title('Legend')
plt.plot([1, 2, 3, 4], label='asc')  # 증가
plt.plot([4, 3, 2, 1], label='dexc') # 감소
plt.legend()
plt.show()
```

### 색상
- 그래프 색상은 plot() 함수에 color 속성을 추가합니다.

```python
# 색상
plt.plot([1, 2, 3], [1, 2, 3], color='red')
plt.show()
```

### X,Y축 이름
- 각 축 이름은 plt.xlabel(), plt.ylabel()을 사용합니다.

```python
# 축이름
plt.plot([1, 2, 3], [1, 2, 3])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### 그래프 선모양

```python
# 선모양 변경
plt.plot([1, 2, 3, 4], color='r', linestyle='--', label='dashed') # 빨간색
plt.plot([4, 3, 2, 1], color = 'g', linestyle = ':', label='dotted') # 감소
plt.legend()
plt.show()
```

### 그림 범위 지정
- 그림의 범위를 수동으로 지정하려면 xlim 명령과 ylim 명령을 사용합니다.
<br>
<br>
- 이 명령들은 그림의 범위가 되는 x축, y축의 최소값과 최대값을 지정합니다.

```python
# 그림 범위
plt.title("x, y Range")
plt.plot([10, 20, 30 , 40], [1, 4, 9, 16],
        c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

## 선그래프
- 선그래프(line plot)는 '연속하는 데이터 값'들을 직선이나 곡선으로 연결하여 데이터 값 사이 관계를 나타냅니다.
<br><br>
- 선 그래프를 그리는 plot 함수에 입력할 x, y 축 데이터를 선택합니다.
<br><br>
- 선 그래프는 x축에 시간 변수를 지정하여 시간 변화량을 나타내는 시계열 데이터를 표현하는 그래프로도 많이 사용됩니다.

```python
# 관련 라이브러리 불러오기
import pandas as pd  #판다스
import matplotlib as mpl  #맷플롭립
import matplotlib.pylab as plt  #맷플롭립
import seaborn as sns  #시본
import numpy as np  #넘파이

#데이터 준비
tips = sns.load_dataset("tips")  # 팁 데이터

#데이터 보기
tips

# 단순 그래프
plt.plot(tips.total_bill)
```

## 막대그래프
- 막대그래프(Bar plot)는 '집단 별 차이를 표현'할 때 사용합니다.
<br><br>
- 막대그래프는 데이터 값의 크기에 비례하여 높이를 갖는 직사각형 막대로 표현하며 막대 높이의 상대적 길이 차이를 통해 값의 크기를 시각화할 수 있습니다.<br><br>
- 막대그래프는 plt.bar([x축 데이터], [y축 데이터])를 지정합니다.

```python
#X축 범주형, Y축 연속형
plt.bar(tips.sex, tips.total_bill)
```

## 산점도
- 산점도(Scatter plot)는 서로 다른 '두 연속형 변수' 사이 관계를 나타내며 연속형 변수의 상관성을 확인할 수 있습니다.

```python
#X축 연속형, Y축 연속형
plt.scatter(tips.total_bill, tips.tip)
```

## 히스토그램
- 히스토그램(Histogram)은 '변수가 하나인 데이터의 빈도수'를 그래프로 표현합니다.
<br><br>
- X축과 같은 크기의 여러 구간으로 나누고 각 구간에 속하는 데이터 값의 빈도를 y축에 표시합니다.

```python
#X축 변수, Y축 빈도
plt.hist(tips.total_bill)
plt.hist(tips.time)
```

## 막대그래프
- 시본에서 사용하는 막대그래프는 sns.barplot()으로 파라미터 3개로 범주별 그룹까지 쉽게 표현할 수 있습니다.
<br><br>
- 시본은 막대그래프를 사용할 때 각 막대에 기본적으로 오차 막대(error bar)를 표시합니다.
<br><br>
- 이 오차 막대를 그리는 범위를 '부트 스트랩 신뢰구간(Bootstrap confidenece interval)'이라고 하며 '이 데이터를 기반으로 유사한 상황의 95%가 해당 범위 내에서 결과를 얻을 것"이라는 의미입니다.
<br><br>
- 만약 이 오차 막대에 신뢰 구간이 아니라 표준편차를 표현하고 싶다면 ci="sd" 파라미터를 지정하면 됩니다.

- 변수
sns.barplot(X, Y, data)
X축 데이터프레임 변수, Y축 데이터프레임 변수, data : 데이터프레임
<br><br>
- 단일변수
sns.catplot('변수', kind='count', 데이터프레임)

```python
#평균치 빠르게 집계해서 시각화하기
sns.barplot(x="sex", y="tip", data=tips)
```

```python
#데이터 분석은 주로 평균을 보지만 때로는 다른 것들을 보고 싶을 수도 있습니다. 예를 들면 데이터에 극단치가 많은 경우 중앙값을 확인하거나 데이터가 범주형인 경우 설문 조사 응답과 같이 각 카테고리가 몇 번 표시되는지 계산할 수 있습니다. 시본은 원하는 집계를 매우 쉽고 간편하게 해낼 수 있습니다.
#estimator를 지정하면 됩니다.(여기에 사용할 값으로는 list에 해당하는 함수를 사용할 수 있습니다)

sns.barplot(x= "sex", y="tip", estimator=len, data=tips)
```

```python
# 여려 열에서 집단별로 그룹하고 세부 집단 시각화 하기
# hue 파라미터 추가
sns.barplot(x="sex", y="tip", hue="day", data=tips)
```

```python
#표준편차 오차막대
sns.barplot(x="sex", y="tip", hue="day", data=tips, ci="sd")
```

```python
#요일별 팁 카운트
sns.catplot('day', kind='count', data=tips)
```

## 박스플롯
- 박스플롯의 정확한 명칭은 상자 수염 그림(box-and-whisker plot)입니다.
<br><br>
- 존 터키(John W. Tukey)가 데이터 분포를 표현하기 위한 시각적 방법으로 만든 탐색적 그래프입니다.

>- 최소값 : 제 1사분위에서 1.5 IQR(Inter Quartile Range)을 뺀 위치입니다.
>- 제 1사분위(Q1) : 25% 위치 값
>- 제 2사분위(Q2) : 50% 위치로 중앙값(median) 값
>- 제 3사분위(Q3) : 75%의 위치를 의미한다.
>- 최대값 : 제 3사분위에서 1.5 IQR을 더한 위치입니다.

```python
sns.boxplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(data = tips, orient="h")
```

## 박스플롯으로 이상치 제거 후 시각화
